```python
import os
import PyPDF2
import logging

# Setup logging
logging.basicConfig(level=logging.INFO, format='%(levelname)s: %(message)s')

# Define the input and output directories
input_dir = '/mnt/data/skanda/MSc_IRD_LLM/data/pdf_data'
output_dir = os.path.join(os.path.dirname(input_dir), 'txt_data')

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Iterate through all files in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith('.pdf'):
        pdf_path = os.path.join(input_dir, filename)
        txt_filename = os.path.splitext(filename)[0] + '.txt'
        txt_path = os.path.join(output_dir, txt_filename)
        
        try:
            # Open the PDF file
            with open(pdf_path, 'rb') as pdf_file:
                pdf_reader = PyPDF2.PdfReader(pdf_file)
                
                # Extract text from each page
                text = ''
                for page in pdf_reader.pages:
                    page_text = page.extract_text()
                    if page_text:
                        text += page_text
                
                # Write the extracted text to a new TXT file
                with open(txt_path, 'w', encoding='utf-8') as txt_file:
                    txt_file.write(text)
            
            logging.info(f"Converted {filename} to {txt_filename}")
        
        except Exception as e:
            logging.error(f"Failed to convert {filename}: {e}")

logging.info("Conversion complete. TXT files saved in the 'txt_data' folder.")
```
